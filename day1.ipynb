```python
class Atm:
    def __init__(self):
        self.balance = 0
        self.pin = ''
        self.menu()

    def menu(self):
        user_input = input("""
        1. Enter 1 to Check Balance
        2. Enter 2 to Deposit
        3. Enter 3 to Withdraw
        4. Enter 4 to Change PIN
        5. Enter 5 to add Pin/Balance
        6. Enter anything else to Exit
        """)

        if user_input == '1':
            self.checkBal()
        elif user_input == '2':
            self.deposit()
        elif user_input == '3':
            self.withdraw()
        elif user_input == '4':
            self.changePin()
        elif user_input == '5':
            self.addPB()
        else:
            print("Exiting...")
            exit()

    def addPB(self):
        user_pin = input("Enter your pin: ")
        self.pin = user_pin

        user_balance = input("Enter your balance: ")
        self.balance = user_balance

        print("Pin and Balance added successfully.")
        self.menu()

    def checkBal(self):
        print("Your current balance is: ", self.balance, "Rs")
        self.menu()

    def changePin(self):
        if self.pin != '':
            ctr = 0
            check = input("Enter your existing pin: ")
            if check == self.pin:
                user_pin = input("Enter new pin: ")
                self.pin = user_pin
                print("Pin changed successfully.")
                self.menu()
            else:
                print("Incorrect pin. Please try again.")
                self.changePin()
                ctr += 1
                if ctr == 3:
                    print("Too many incorrect attempts. Exiting to menu.")
                    self.menu()
        else:
            print("No pin set. Please add a pin first.")
            self.addPB()

    def withdraw(self):
        if self.pin != '':
            user_pin = input("Enter your pin: ")
            if user_pin == self.pin:
                amount = float(input("Enter amount to withdraw:"))
                if amount <= self.balance and self.balance > 0:
                    self.balance -= amount
                    print("Amount withdrawn successfully.  Balance:", self.balance, "Rs")
                    self.menu()
                else:
                    print("Insufficient balance.")
                    self.withdraw()
            else:
                print("Incorrect pin.")
                self.withdraw()
        elif self.balance == 0:
            print("Your account is empty.")
            self.menu()
        else:
            print("No pin set. Please add a pin first.")
            self.addPB()

    def deposit(self):
        if self.pin != '':
            ctr = 0
            user_pin = input("Enter your pin: ")
            if user_pin == self.pin:
                amount = float(input("Enter amount to deposit:"))
                self.balance += amount
                print("Money added successfully. Balance:", self.balance, "Rs")
                self.menu()
            else:
                print("Incorrect pin.")
                ctr+=1
                if ctr == 3:
                    print("Too many incorrect attempts. Exiting to menu.")
                    self.menu()
        else:
            print("No pin set. Please add a pin first.")
            self.addPB()
```
